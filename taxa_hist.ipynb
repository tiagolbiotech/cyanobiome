```python
import glob
import re
import os
from collections import Counter
import csv
```

```python
glob_list = glob.glob('./ncbi_genomes/*.fasta')

glob_list
```

```python
taxa_list = []

for filepath in glob_list:
    if os.stat(filepath).st_size != 0:
        print(filepath)
        with open(filepath) as f:
            first_line = f.readline()
            print(first_line.split(' '))
            if 'TPA_asm:' in str(first_line):
                taxa = first_line.split(' ')[2]
            if 'uncultured' in str(first_line):
                taxa = first_line.split(' ')[2]
            if 'TPA_asm:' not in str(first_line) and 'uncultured' not in str(first_line):
                taxa = first_line.split(' ')[1]
            taxa_list.append(taxa)
```

```python
count_dict = Counter(taxa_list)

count_dict
```

```python
w = csv.writer(open("taxa_hist_dict.csv", "w"))
for key, val in count_dict.items():
    w.writerow([key, val])
```
